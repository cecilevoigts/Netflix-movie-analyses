```python
#creating a set of genres that are in both movies and TV shows
set_types = set(types)
set_types_movies = set(types_movies)
set_both = set_types.intersection(set_types_movies)
print(set_both)
```

```python
def certain_genres(genres, counts, filter_set):
    dic = {}
    for genre, count in zip(genres, counts):
        if genre in filter_set:
            dic[genre] = count
    return dic

certain_genres_movies = certain_genres(df_genres_movies['genre'], df_genres_movies['count'], set_both)
print(certain_genres_movies)
```

```python
certain_genres_shows = certain_genres(df_genres_shows['genre'], df_genres_shows['count'], set_both)
print(certain_genres_shows)
```

```python
certain_movies = pd.DataFrame({'Movies': certain_genres_movies.values()}, index=certain_genres_movies.keys())
certain_movies = certain_movies.reset_index()
#display(certain_movies)

certain_shows = pd.DataFrame({'TV Shows': certain_genres_shows.values()}, index=certain_genres_shows.keys())
certain_shows = certain_shows.reset_index()
#display(certain_shows)

#merge both dfs
both = pd.merge(certain_movies, certain_shows, on='index', how='left')
display(both)
```

```python
#reshaping the df to better plot it
both_melted = pd.melt(both, id_vars='index', value_vars=['Movies', 'TV Shows'], var_name='Type', value_name='Count')
display(both_melted)
```

```python
both_plot = sns.barplot(data=both_melted, x='index', y='Count', hue='Type', palette=["#ff0100", "#0032c6"])
both_plot.set_xticklabels(labels=both_melted['index'], rotation=45, ha='right')
both_plot.set(xlabel='Genre', title='Amount of Movies vs TV Shows per Genre')
both_plot.legend(labelcolor='w')
plt.savefig('Genres both.jpg', bbox_inches='tight')
```

```python
#cummulative graph
#pct for movies
pct_movies=[]
all_m = both['Movies'].sum()
for value in both['Movies']:
    pct = value / all_m * 100
    pct_movies.append(pct)

#pct for shows
pct_shows=[]
all_s = both['TV Shows'].sum()
for value in both['TV Shows']:
    pct_s = value / all_s * 100
    pct_shows.append(pct_s)

both['movies'] = pct_movies
both['tv shows'] = pct_shows    
display(both)

both_melted_pct = pd.melt(both, id_vars='index', value_vars=['movies', 'tv shows'], var_name='Type', value_name='Pct')
display(both_melted_pct)
```

```python
both_pct = sns.barplot(data=both_melted_pct, x='index', y='Pct', hue='Type', palette=["#ff0100", "#0032c6"])
both_pct.set_xticklabels(labels=both_melted['index'], rotation=45, ha='right')
both_pct.set(xlabel='Genre', title='Amount of Movies vs TV Shows per Genre in %')
both_pct.legend(labelcolor='w')
plt.savefig('Genres both pct.jpg', bbox_inches='tight')
plt.ylabel('Percent')
```

```python
#specifying which ratings are what age group
groups = {'G':'Children', 'NC-17':'Adults', 'NR':'Adults', 'PG':'Older Children', 'PG-13':'Teens', 'R':'Adults', 'TV-14':'Teens', 'TV-G':'Children', 'TV-MA':'Adults', 'TV-PG':'Older Children', 'TV-Y':'Children', 'TV-Y7':'Older Children', 'TV-Y7-FV':'Older Children', 'UR':'Adults'}

age_groups = []
for rating in df['rating']:
    age_group = groups[rating]
    age_groups.append(age_group)

df['age_group'] = age_groups
display(df.head(10))
```

```python
ages = df.groupby('age_group').count().reset_index().sort_values(by='age_group', ascending=False)
display(ages)
```

```python
#donut pie chart
colours=["#fdebcc", "#bb0001", "#ff0100", "#0032c6"]
explode = (0, 0, 0, 0.08)
age, text, autotext = plt.pie(ages['show_id'], labels=ages['age_group'], autopct='%1.1f%%', startangle=115, explode=explode, pctdistance=.8, colors=colours, wedgeprops={'linewidth':0})
plt.title('What age group is the most content made for?')

#setting only one label as bold
adults = 3
text[adults].set_fontweight('bold')
autotext[adults].set_fontweight('bold')

#making the donut
my_circle=plt.Circle((0,0), 0.6, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.savefig('Age groups.jpg', dpi=300, bbox_inches='tight')
```
