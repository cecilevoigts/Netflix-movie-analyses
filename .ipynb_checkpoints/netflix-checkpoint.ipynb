```python
#importing the csv file as a df
import pandas as pd
df = pd.read_csv('netflix_titles.csv')
display(df.head(15))
```

```python
#first looking at the genre variety
df_genre = df.groupby('listed_in')
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
genre = sns.histplot(data=df, x='listed_in', multiple='dodge', shrink=.8)
```

```python
type = sns.histplot(data=df, x='type')
```

```python
release = sns.histplot(data=df, x='release_year')
```

```python
#genre for TV Shows
mask_show = df['type'] == 'TV Show'
df_series = df[mask_show]

df_series['cleaned_genre'] = df_series['listed_in'].str.replace(' TV Shows', '').str.replace('TV ', '').str.replace(' TV', '')

display(df_series.head(15))
```

```python
series_genre = sns.histplot(data=df_series, x='cleaned_genre')
series_genre.set_xticklabels(labels=df_series['cleaned_genre'], rotation=45)
```

```python
#defining a function that counts all of the genres
def count_genre(genre):
    count = df_series['cleaned_genre'].str.contains(genre).sum()
    return count
```

```python
dramas = count_genre('Dramas')
print(dramas)
```

```python
comedies = count_genre('Comedies')
print(comedies)
```

```python
#putting all of the genres and their count in a dictionary
genres = {}
types = ['Dramas', 'Comedies', 'International', 'Mysteries', 'Crime', 'Action & Adventure', 'Docuseries', 'Reality', 'Romantic', 'Horror', 'British', 'Spanish-Language', 'Thrillers', 'Kids', 'Sci-Fi & Fantasy', 'Anime Series', 'Korean', 'Teen', 'Shows', 'Stand-Up Comedy & Talk Shows', 'Classic & Cult', 'Science & Nature']
for genre in types:
    sum = count_genre(genre)
    genres[genre] = int(sum)

print(genres)
```

```python
import numpy as np
ind = np.arange(len(genres))
df_genres_shows = pd.DataFrame(genres.items(), columns=['genre', 'count'] ,index=ind)
display(df_genres_shows)
```

```python
#making a plot using the dictionary above
sns.set_style("whitegrid")

shows_genre = sns.barplot(data=df_genres_shows, x='genre', y='count')
shows_genre.set_xticklabels(labels=df_genres_shows['genre'], rotation=90)
shows_genre.set(xlabel ="Genres", ylabel = "Count", title ='Number of TV show genres')
```
